```python
import os
import pandas as pd
import glob

Root = 'D:/Actor_01/'

file_emotion = []
file_path = []

# List all WAV files inside the "D:/Actor_01" directory
actor_files = glob.glob(os.path.join(Root, '*.wav'))
for file in actor_files:
    part = os.path.basename(file).split('.')[0]
    part = part.split('-')
    # Third part in each file represents the emotion associated with that file.
    file_emotion.append(int(part[2]))
    file_path.append(file)

# DataFrame for emotion of files
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])

# DataFrame for path of files.
path_df = pd.DataFrame(file_path, columns=['Path'])
Ravdess_df = pd.concat([emotion_df, path_df], axis=1)

# Changing integers to actual emotions.
Ravdess_df.Emotions.replace({
    1: 'neutral',
    2: 'calm',
    3: 'happy',
    4: 'sad',
    5: 'angry',
    6: 'fear',
    7: 'disgust',
    8: 'surprise'
}, inplace=True)

# Display all the rows of the DataFrame.
print(Ravdess_df)
```

```python
import os

# Define the folder path
folder_path = 'D:/Actor_01'

# Define a dictionary to map emotion labels to their corresponding emotions
emotion_mapping = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fear',
    '07': 'disgust',
    '08': 'surprise'
}

# Lists to store emotion labels and corresponding emotions
emotion_labels = []
emotions = []

# Loop through all files in the folder
for file_name in os.listdir(folder_path):
    # Check if the file is a .wav file
    if file_name.lower().endswith('.wav'):
        # Split the file name by '-' to extract the emotion label (3rd index)
        parts = file_name.split('-')
        emotion_label = parts[2]

        # Map the emotion label to its corresponding emotion
        emotion = emotion_mapping.get(emotion_label)
        
        # Append the emotion label and emotion to their respective lists
        emotion_labels.append(emotion_label)
        emotions.append(emotion)

# Print the emotion labels and corresponding emotions
print("Emotion Labels:", emotion_labels)
print("Emotions:", emotions)
```
