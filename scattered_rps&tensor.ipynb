```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import wavfile

# Read the audio file
sampling_rate, audio_data = wavfile.read("D:/wav/13a02Ad.wav")

# Normalize audio data to the range [-1, 1]
audio_data = audio_data.astype(np.float32) / np.max(np.abs(audio_data))

# Embedding parameters
embedding_dimension = 3  # Choose the embedding dimension
time_delay = 17          # Choose the time delay

# Create the embedded time series matrix
def create_embedded_matrix(data, embedding_dimension, time_delay):
    num_points = len(data) - (embedding_dimension - 1) * time_delay
    embedded_matrix = np.zeros((num_points, embedding_dimension))
    for i in range(num_points):
        for j in range(embedding_dimension):
            embedded_matrix[i, j] = data[i + j * time_delay]
    return embedded_matrix

embedded_matrix = create_embedded_matrix(audio_data, embedding_dimension, time_delay)

# Scale the embedded matrix values from [-1, 1] to [0, 256]
min_value = np.min(embedded_matrix)
max_value = np.max(embedded_matrix)
scaled_embedded_matrix = ((embedded_matrix - min_value) / (max_value - min_value)) * 256

# Scatter plot of the scaled embedded points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(scaled_embedded_matrix[:, 0], scaled_embedded_matrix[:, 1], scaled_embedded_matrix[:, 2], s=1, c='green', label='Scattered Line')
ax.set_title("forward-in-time (Scaled to 0-256)")
ax.legend()
plt.show()
```
